```python
train_sub = Path("/root/datasets/masc/train/subtitles")
print(train_sub.exists())
captions_duration = []
for sub in tqdm(train_sub.glob("*"), total=len(list(train_sub.glob("*")))):
    vtt = webvtt.read(sub)
    for caption in vtt:
        captions_duration.append((caption.end_in_seconds - caption.start_in_seconds, caption.text))
caption_df = pd.DataFrame(captions_duration, columns=["duration", "text"])
```
